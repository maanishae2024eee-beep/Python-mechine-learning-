```python
1.Write a Python script for Weather Data Analysis:
Problem Statement:
You have a CSV file named 'weather_data.csv' containing daily weather data for a city.

Date Temperature
(°C)

Humidity
(%)

Wind
Speed
(km/h)

Rainfall
(mm)
01-01-2025 15.2 80 10.5 0
02-01-2025 16.8 75 8 0
03-01-2025 14.5 85 12 1.2
04-01-2025 13 90 15.2 5.4
05-01-2025 17.6 70 7.4 0
06-01-2025 18.1 65 6 0
07-01-2025 16.2 78 9.1 0.5
08-01-2025 12.4 88 14.3 2.1
09-01-2025 13.8 83 11.5 0
10-01-2025 15.5 79 10 0
Perform the following tasks:
a. Check for missing values and handle them appropriately,
b. Calculate the average temperature for each month,
c. Visualize the monthly average temperature using a bar plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file with proper encoding
df = pd.read_csv("weather_data.csv", encoding="latin1")

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Check missing values
print("Missing Values:\n", df.isnull().sum())

# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Extract month
df['Month'] = df['Date'].dt.month

# Calculate average temperature for each month
avg_temp = df.groupby('Month')['Temperature (°C)'].mean()

print(avg_temp)

# Bar plot
avg_temp.plot(kind='bar')
plt.title("Monthly Average Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()
```

```python
2. Write a Python script using pandas for Student Performance Analysis:
Problem Statement:
You are given a CSV file named 'student_scores.csv' containing scores of students in
various subjects.

Student
ID Name Maths Physics Chemistry
101 Alice Brown 85 88 87
102 Ben Carter 78 75 80
103 Clara Davis 92 94 90
104 David Evans 65 68 66
105 Eva Foster 74 78 76
106 Frank Green 60 55 58
107 Grace Hall 88 85 86
108 Henry Ives 70 72 74
109 Isla Jones 95 96 94
110 Jack King 82 80 78

Perform the following tasks:
a. Display the summary statistics for each subject,
b. Calculate the average score for each student,
c. Identify students who scored below 60 in more than two subjects.
```

```python
import pandas as pd

# Create DataFrame directly
data = {
    'Student ID': [101,102,103,104,105,106,107,108,109,110],
    'Name': ['Alice Brown','Ben Carter','Clara Davis','David Evans',
             'Eva Foster','Frank Green','Grace Hall','Henry Ives',
             'Isla Jones','Jack King'],
    'Maths': [85,78,92,65,74,60,88,70,95,82],
    'Physics': [88,75,94,68,78,55,85,72,96,80],
    'Chemistry': [87,80,90,66,76,58,86,74,94,78]
}

df = pd.DataFrame(data)

# Summary statistics
print("Summary Statistics:")
print(df[['Maths','Physics','Chemistry']].describe())

# Average marks
df['Average'] = df[['Maths','Physics','Chemistry']].mean(axis=1)

print("\nAverage Score of Each Student:")
print(df[['Student ID','Name','Average']])

# Below 60 in more than two subjects
low_score = df[
    ((df['Maths'] < 60).astype(int) +
     (df['Physics'] < 60).astype(int) +
     (df['Chemistry'] < 60).astype(int)) > 2
]

print("\nStudents scoring below 60 in more than two subjects:")
print(low_score[['Student ID','Name']])
```

```python
3. Write a Python script for Flight Data Analysis:
Problem Statement:
You have a CSV file named 'flight_data.csv' containing flight information such as origin,
destination, and departure time.
Perform the following tasks:

a. Display the summary statistics for departure delays,
b. Calculate the average delay for each airline,
c. Identify the most common departure and arrival destinations,
d. Visualize the distribution of departure delays using a box plot.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame directly
data = {
    'Airline': ['IndiGo', 'Air India', 'SpiceJet', 'IndiGo', 'Vistara',
                'Air India', 'SpiceJet', 'IndiGo', 'Vistara', 'Air India'],
    'Origin': ['Chennai', 'Delhi', 'Mumbai', 'Chennai', 'Bangalore',
               'Delhi', 'Mumbai', 'Chennai', 'Bangalore', 'Delhi'],
    'Destination': ['Mumbai', 'Chennai', 'Delhi', 'Hyderabad', 'Delhi',
                    'Mumbai', 'Chennai', 'Delhi', 'Mumbai', 'Bangalore'],
    'Departure Delay': [15, 30, 10, 5, 20, 25, 12, 8, 18, 35],
    'Arrival Delay': [10, 25, 8, 3, 15, 20, 10, 5, 12, 30]
}

df = pd.DataFrame(data)

# a. Summary statistics for departure delays
print("Summary Statistics for Departure Delays:")
print(df['Departure Delay'].describe())

# b. Average delay for each airline
print("\nAverage Departure Delay for Each Airline:")
print(df.groupby('Airline')['Departure Delay'].mean())

# c. Most common origin and destination
print("\nMost Common Origin:")
print(df['Origin'].mode()[0])

print("\nMost Common Destination:")
print(df['Destination'].mode()[0])

# d. Box plot for departure delays
plt.boxplot(df['Departure Delay'])
plt.title("Distribution of Departure Delays")
plt.ylabel("Delay (minutes)")
plt.show()
```

```python
import math

# Euclidean Distance
x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("Euclidean Distance =", distance)

# Dot Product of Two Matrices
r1 = int(input("Enter rows of Matrix A: "))
c1 = int(input("Enter cols of Matrix A: "))

A = []
print("Enter Matrix A:")
for i in range(r1):
    row = list(map(int, input().split()))
    A.append(row)

r2 = int(input("Enter rows of Matrix B: "))
c2 = int(input("Enter cols of Matrix B: "))

B = []
print("Enter Matrix B:")
for i in range(r2):
    row = list(map(int, input().split()))
    B.append(row)

if c1 == r2:
    result = []

    for i in range(r1):
        row = []
        for j in range(c2):
            s = 0
            for k in range(c1):
                s += A[i][k] * B[k][j]
            row.append(s)
        result.append(row)

    print("Dot Product:")
    for i in result:
        print(i)
else:
    print("Matrix multiplication not possible")
```

```python
5. Write a program to work with time series data using NumPy and pandas
Problem Statement: A time series dataset representing daily temperatures over a year.
You will perform some basic analysis and visualization using NumPy, pandas, and
matplotlib.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate daily temperatures for one year
dates = pd.date_range("2025-01-01", periods=365, freq="D")
temps = np.random.randint(20, 40, 365)

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Temperature": temps
})

# Set Date column as index
df.set_index("Date", inplace=True)

# Basic Analysis
print("First 5 Records:")
print(df.head())

print("\nSummary Statistics:")
print(df.describe())

print("\nMonthly Average Temperature:")
print(df.resample("M").mean())

# Line Plot
plt.figure(figsize=(10,5))
plt.plot(df.index, df["Temperature"])
plt.title("Daily Temperature Over One Year")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()

# Bar Plot for Monthly Average
df.resample("M").mean().plot(kind="bar", figsize=(10,5))
plt.title("Monthly Average Temperature")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
6. Write a program to simulate and analyze daily stock prices over one year and generate
the stock prices, calculate some basic metrics, and visualize the data with additional
features like moving averages and visualize the same.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate stock prices for one year
np.random.seed(10)
dates = pd.date_range("2025-01-01", periods=365, freq="D")
prices = 100 + np.cumsum(np.random.randn(365))

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Stock Price": prices
})

# Set Date as index
df.set_index("Date", inplace=True)

# Basic Metrics
print("First 5 Records:")
print(df.head())

print("\nSummary Statistics:")
print(df.describe())

print("\nMaximum Price =", df["Stock Price"].max())
print("Minimum Price =", df["Stock Price"].min())
print("Average Price =", df["Stock Price"].mean())

# Moving Averages
df["MA_7"] = df["Stock Price"].rolling(window=7).mean()
df["MA_30"] = df["Stock Price"].rolling(window=30).mean()

# Plot Stock Price and Moving Averages
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Stock Price"], label="Stock Price")
plt.plot(df.index, df["MA_7"], label="7-Day Moving Avg")
plt.plot(df.index, df["MA_30"], label="30-Day Moving Avg")
plt.title("Daily Stock Prices Over One Year")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
7. Write a simple Python program using pandas that creates a DataFrame, performs the
following basic operations, and prints the result.
a. Creates two lists: data containing fruit names and prices containing their
corresponding prices. Zips these lists together and creates a DataFrame named
fruits_df with columns named "Fruit" and "Price".
b. Uses info() to get information about the DataFrame, including data types and
number of entries.
c. Prints the entire DataFrame using to string().
d. Calculates descriptive statistics (mean, standard deviation, etc.) for the "Price"
column and prints the results
```

```python
import pandas as pd

# a. Create lists
data = ["Apple", "Banana", "Orange", "Mango", "Grapes"]
prices = [120, 40, 80, 150, 90]

# Create DataFrame
fruits_df = pd.DataFrame(list(zip(data, prices)), columns=["Fruit", "Price"])

# b. DataFrame info
print("DataFrame Info:")
fruits_df.info()

# c. Print entire DataFrame
print("\nComplete DataFrame:")
print(fruits_df.to_string(index=False))

# d. Descriptive statistics for Price column
print("\nDescriptive Statistics for Price:")
print(fruits_df["Price"].describe())
```
