```python
#open a text file in'write' mode to write given lines
f=open("D:\textfile.txt",'w')
f.write("first line\n")
f.write("second line\n")
f.write("third line\n")
f.close
#open a text file in'Read' mode to read and print the available lines-read functions
k=open("D:\textfile.txt",'r')
print(k.readline())
print(k.readline())
print(k.readline())
k.close()
```

```python
#open a text file in'Append' mode to addd\apppend new line inputs
m=open("D:\textfile.txt",'a')
m.write("fourth line\n")
m.write("fifth line\n")
m.close()
#open a text file in'Read' mode to read and print the available full text - read function
p=open("D:\textfile.txt",'r')
print(p.read())
p.close()
```

```python
with open("D:\textfile.txt", "r") as file:
    lines = file.readlines()
    print(lines)
```

```python
import json
#JSON string
x='{"name":"john","age":30,"City":"New York"}'

#convert JSON
y=json.loads(x)
print(y["age"])
```

```python
#Python dictionary
data={
    "name":"john",
    "age":30,
    "City":"New York"
}

#convert python dictionary to json string
json_data=json.dumps(data)
print(json_data)
```
