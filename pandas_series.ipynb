```python
import pandas as pd
import numpy as np

# Creating Series from Array
arr = np.array(['P', 'a', 'n', 'd', 'a', 's'])
a = pd.Series(arr)
print("Series from array:")
print(a)

# Series from dictionary
arr = {'x': 0., 'y': 1., 'z': 2.}
b = pd.Series(arr)
print("\n\nSeries from dictionary:\n")
print(b)
```

```python
x = pd.Series(4, index=[0, 1, 2, 3])
print("\nSeries using scalar\n")
print(x)
```

```python
y = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
print("\nSeries through index:")
print(y)
```

```python
z = pd.Series(data=[1, 2, 3, 4])
print("\n a series:\n", z)
print("\nIndex: \n", z.index)
print("\nValues: \n", z.values)
print("\nShape: ", z.shape)
print("\nDimension: ", z.ndim)
print("\nSize: ", z.size)
```
