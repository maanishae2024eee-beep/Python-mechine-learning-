```python
#Empty dataframe
import pandas as pd
df=pd.DataFrame()
print(df)
```

```python
#Series to Data Frame
emp=pd.Series(['Paker','John','Smith','William'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\nSeries to Data Frame\n")
print(result)
```

```python
#Extraction one column
print("\nExtraction one column:\n")
print(result['Emp'])
```

```python
#Adding a one extra column
print("\nAdding new column:\n")
result['Age']=pd.Series([35,24,40,38])
print(result)
```

```python
#Deleting one column
print("\nDeleting one column:\n")
del result['Age']
print(result)
```

```python
#Extracting the third rows
print("\nExtracting the third rows:\n")
print(result.loc[2])
```

```python
#slice rows
print("\nSlice rows:\n",result[1:3])
```

```python
#Adding new row values
d2=pd.DataFrame([['Dale,123'],['Mark',143]],columns=['Emp','ID'])
result=pd.concat([result,d2],ignore_index=True)
print("\nAdding new row values:\n",result)
```

```python
#Deleting the particular row
print("\nDeleting particular row:\n",result.drop(1))
```
