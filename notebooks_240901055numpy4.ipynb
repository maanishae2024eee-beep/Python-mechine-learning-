```python
import numpy as np
arr1=[10,20,30,40,50]
arr2=[2,4,5,8,10]
a=np.array(arr1)
b=np.array(arr2)
print("original arrays")
print(a)
print(b)
print("\n vector addition")
print(a+b)
print("\n vector subtraction")
print(a-b)
print("\n vector multiplication")
print(a*b)
print("\n vector division")
print(a/b)
print("\n vector dot product")
print(a.dot(b))
print("\n scalar multiplication")
sclr=5
print("scalar value=",sclr)
```

```python
#numpy.Vectorize method
def my_func(x,y):
    #return x-y if x>y,otherwise return x+y
    if x>y:
        return x-y
    else:
        return x+y
print("\n\n Numpy.vectorize method")
print("(Return x-y if x>y,otherwise return x+y)")
arr1=[10,4,20]
arr2=[2,3,30]
vec_func=np.vectorize(my_func)
print("array1:",arr1)
print("array2:",arr2)
print("result:",vec_func(arr1,arr2))
```
